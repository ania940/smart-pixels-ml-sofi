```python
import matplotlib.pyplot as plt
import numpy as np
import mplhep as hep
import json
```

```python
import matplotlib.colors as mcolors
colors = list(mcolors.TABLEAU_COLORS.keys()) + list(mcolors.BASE_COLORS.keys())
```

```python
with open('params.json') as f:
    params = json.load(f)
```

```python
fig, ax = plt.subplots()

n_models = len(params.keys())
print(n_models)
ax.set_xlim(-1,n_models)
ax.set_xticks([])
ax.set_ylabel("Number of trainable parameters")

x_array = np.linspace(0,n_models-1,n_models)
y_array = list(params.values())

print(x_array,y_array)

dots = []
for i in range(0,n_models):
    dots += [ax.scatter(x_array[i], y_array[i], marker='o', color = colors[i])]

#lines = []
labels = list(params.keys())
    
ax.legend(dots,labels=labels, bbox_to_anchor=(1.4, 1))
plt.show()
```
